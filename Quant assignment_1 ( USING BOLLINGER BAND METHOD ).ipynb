```python
df['30_MA_value']=df['value'].rolling(window=30).mean()
df['20_std_value']=df['value'].rolling(window=20).std()
```

```python
df['Upper']=df['30_MA_value']+ 1.8*df['20_std_value']
df['Lower']=df['30_MA_value']-1.8*df['20_std_value']
```

```python
df[['value','30_MA_value','Upper','Lower']].plot(figsize=(30,50))
```
